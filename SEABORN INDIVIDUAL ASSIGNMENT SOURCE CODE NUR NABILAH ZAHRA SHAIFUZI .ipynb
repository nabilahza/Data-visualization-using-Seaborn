### Dataframes of Loan datasets


```python
loan=pd.read_csv('Loan.csv')
loan.head()
```

### Check if datasets consist of null value or not

```python
loan.apply(lambda x: sum(x.isnull()),axis=0)
```

### Remove null value from the datasets

```python
loan=loan.dropna()
loan.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
loan
```

### Remove columns in datasets that will not be used 

```python
loan= loan.drop('Gender', axis=1)
loan
```

```python
loan= loan.drop('Dependents', axis=1)
loan
```

```python
loan= loan.drop('Married', axis=1)
loan
```

```python
loan= loan.drop('CoapplicantIncome', axis=1)
loan
```

```python
loan= loan.drop('Loan_Amount_Term', axis=1)
loan
```

```python
loan= loan.drop('LoanAmount', axis=1)
loan
```

### Rename Y and N for Loan_Status column

```python
mapping={'Y': 'Yes', 'N': 'No'}
loan['Loan_Status']=loan['Loan_Status'].replace(mapping)
loan
```

### Count Plot for Each Category that Have Been Set As Goals
